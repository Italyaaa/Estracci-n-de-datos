```python
#Importar las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Estocolmo.csv")
df
```

```python
#Los registros de las propuedades que sólo sean "Entire loft" y "Entire guesthouse"
PT = df[(df["property_type" ] == "Entire loft") | (df["property_type"] == "Entire guesthouse")]
#PT.to_csv("PuntoA.CSV")
PT
```

```python
#Los registros de los host  que se hayan unido a Airbnb, después del año 2020
df["host_since"] = df["host_since"].astype('datetime64[s]')
HS_2020 = df[(df["host_since"] >= '01-01-20')]
#HS_2020.to_csv("PuntoB.CSV")
HS_2020
```

```python
#Host que respondan a más tardar en un día
HRT = df[(df["host_response_time" ] == "within a day") | (df["host_response_time"] == "within a few hours") | (df["host_response_time"] == "within an hour")]
#HRT.to_csv("PuntoC.CSV")
HRT
```

```python
#Registros de las filas múltiplos de 200
iM200 = df.index[df.index % 200 == 0]
iM200
i200= df.iloc[ [200, 400, 600, 800, 1000, 1200, 1400, 1600, 1800, 2000, 2200, 2400, 2600, 2800, 3000, 3200, 3400, 3600, 3800, 4000, 4200, 4400, 4600, 4800, 5000, 5200], : ]
#i200.to_csv("PuntoD.CSV")
i200
```

```python
#Registros de los tipos de cuarto "Entire home" y "Private room"
TC = df[(df["room_type" ] == "Entire home") | (df["room_type"] == "Private room")]
#TC.to_csv("PuntoE.CSV")
TC
```

```python
#Registros que "a lo mucho cuenten con 4 camas"
menos4camas = df[(df["beds"] <= 4)]
#menos4camas.to_csv("PuntoF.CSV")
menos4camas
```

```python
#Registros que sean "super host"
SH= df[(df["host_is_superhost" ] == "t")]
#SH.to_csv("PuntoG.CSV")
SH
```

```python
#Registro de las columnas impares
Ci = df.columns[::2]
Cimpares = df[(Ci)]
#Cimpares.to_csv("PuntoH.CSV")
Cimpares
```

```python
df.info()
```

```python
df['host_response_rate']= df['host_response_rate'].astype(float)
```

```python
df['host_response_rate']=df['host_response_rate'].str.replace('%', '')
```

```python
#Los registros cuyo "host response" sea mayor a 95% y que no sea "super host"
HR = df[(df["host_response_rate"] >= 95) & (df["host_is_superhost"] == "f")]
#HR.to_csv("PuntoA.2.CSV")
HR
```

```python
#Baño no igual a 1 
NO1B = df[~(df["bathrooms"] == 1)]
#NO1B.to_csv("PuntoB.2.CSV")
NO1B
```

```python
df["price"]=df["price"].str.replace("$", "")
```

```python
df["price"]=df["price"].str.replace(',','')
df["price"]=df["price"].str.replace(' ','')
```

```python
df["price"]=df["price"].astype(float)
```

```python
#precio entre 100 y 200
ECYD = df[(df["price"] <= 200) & (df["price"] >100)]
#ECYD.to_csv("PuntoC.2.CSV")
ECYD
```

```python
#Registros que sea "instant bookable" y hayan registrado su perfil
PD = df[(df["instant_bookable"] == "t") & (df["host_identity_verified"] == "t")]
#PD.to_csv("PuntoD.2.CSV")
PD
```

```python
#Tipos de cuarto "hotel room" y "private room"
PE = df[(df["room_type"] == "Private room") | (df["room_type"] == "hotel room")]
PE.to_csv("PuntoE.2.CSV")
PE
```

```python
#Más de 4 "accommodates"
PF = df[(df["accommodates"] > 4)]
PF.to_csv("PuntoF.2.CSV")
PF
```
